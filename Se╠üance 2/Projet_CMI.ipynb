```python

stations_pdc_annuel.columns = ['Annee','N_pdc','N_stations']
stations_pdc_annuel.head()
```

A FAIRE : décrire cet extrait de code ci-dessus

```python
# calcul des cumuls des nombres de stations et de points de charge 
stations_pdc_cumul = stations_pdc_annuel.copy()
stations_pdc_cumul['N_pdc'] = stations_pdc_cumul.N_pdc.cumsum()
stations_pdc_cumul['N_stations'] = stations_pdc_cumul.N_stations.cumsum()
stations_pdc_cumul
```

## Jointure entre par et stations

```python
# enlever les valeurs manquantes
stations2022_dep = stations2022.dropna(subset=['code_insee_commune']).copy()
```

```python
stations2022_dep['DEP'] = stations2022_dep.code_insee_commune.values
stations2022_dep['DEP'] = stations2022_dep['DEP'].str.replace('A','0')
stations2022_dep['DEP'] = stations2022_dep['DEP'].str.replace('B','0')
stations2022_dep['DEP'] = stations2022_dep['DEP'].astype(int)//1000
```

A FAIRE décrire le code code ci-dessus

Le code ci dessus remplace les lettres présentes dans les codes départements par des chiffres et change le type str par int

```python
# Départements avec le plus de stations 
nb_stations2022_dep = stations2022_dep.value_counts('DEP').reset_index()
nb_stations2022_dep.columns = ['DEP','NB_stations']
```

A FAIRE : creer une variable département "DEP" pour le dataframe parc2022

```python
parc2022['DEP'] = parc2022['INSEE'].str.replace('A','0').str.replace('B','0').astype(int)//1000

parc2022.columns = ['INSEE','NB_VE','DEP']
```

```python
# aggregation au niveau 
parc_ve_dep = parc2022.groupby('DEP')['NB_VE'].sum().reset_index()
```

```python
# pour classer par rapport à une colonne
parc_ve_dep.sort_values('NB_VE',ascending=False)
```

```python
# faire une jointure sur 
parc_stations = pd.merge(parc_ve_dep,nb_stations2022_dep, on = 'DEP')
parc_stations
```

## Analyse croisée entre parc et stations

```python
# graphique de points entre nb_ve et nb_stations
plt.figure( figsize=(8,8) ) # option pour régler la taille de la figure 
plt.scatter(parc_stations.NB_stations, parc_stations.NB_VE)
plt.ylabel('Nombre de véhicules électrique')
plt.xlabel('Nombre de stations')
```

A FAIRE : affichez les légendes correspondantes des axes 

```python
# graphique de points entre nb_ve et nb_stations
plt.scatter(parc_stations.NB_stations, parc_stations.NB_VE)
plt.yscale('log')
plt.xscale('log')
plt.ylabel('Nombre de véhicules électrique')
plt.xlabel('Nombre de stations')
```

```python
### regressions linéaires avec sklearn 

# on cherche à estimer y = beta0 + beta X
y = parc_stations.NB_VE.values
x = parc_stations.NB_stations.values
X = np.vstack([x]).T # on peut éventuellement rajouter des vecteurs pour une regression à N regresseurs

reg = LinearRegression().fit(X, y)

# affichage des résultats de la regression
print('coefficients',reg.coef_ )
print('intercept',reg.intercept_)
print('R2', reg.score(X,y) )

y_regression = reg.predict(X)


# graphique des données avec le modèle linéaire
plt.scatter(x,y, color = 'orange')
plt.plot(x,y_regression, color = 'purple')
```
